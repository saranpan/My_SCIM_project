> Significantly difference in Mean and Standard Deviation between Individual A & B? With our bare eye is no, so let's check the difference between them using Two-sample T-test & One way ANOVA

> We may say that individuals A & B are not insifnificantly different using just our bare eye, however we can better claim this using Statistics method to test the significantly different which including in T-test (for mean) and One-Way ANOVA (for standard deviation)

First, we match A & B in the form of the the following : <br>

<fieldset>
A_follower_count - 	B_follower_count 	,A_following_count - B_following_count,	A_listed_count -	B_listed_count, ...., A_network_feature_3 - B_network_feature_3

T-Test

````python
from scipy import stats

# To test the difference in mean of A & B


print('Step 1: Setup the hypothesis \n\n') 

print('H0: The mean of group A & B are equal \n') 
print('Ha: The mean of group A & B are not equal \n ') 
print('########################################################################\n')

print('Step 2: Test the statistics \n\n') 

k=0

alpha = 0.05    #critical region

for L in range(1,12):

    _ ,p_value= stats.ttest_ind(a=data.iloc[:,L], b=data.iloc[:,L+11])
    L+=1

    if p_value < alpha:     
      print("The null hypothesis is rejected ; p-value = {:.3f}".format(p_value))

    else:
      print("The null hypothesis is accepted ; p-value = {:.3f} \n".format(p_value))
      k += 1

#print('Step 2: Test the statistics \n\n') 

if k == 11:
  print('######################################################################## \n')
  print('Step 3: Conclusion \n', 
        'We conclude that the mean of all features between A & B are equal, since all features does not fall into rejection region \n', '########################################################################')
else:
  print('######################################################################## \n')
  print('########################################################################\n\n','Step 3: Conclusion \n', 
        'We conclude that the mean of all features between A & B are NOT equal, since some of the features does fall into rejection region \n')
````



---



One-Way ANOVA

````python
# To test the difference in S. Deviation of A & B using One-Way ANOVA

print('Step 1: Setup the hypothesis \n\n') 

print('H0: The variance of group A & B are equal \n') 
print('Ha: The variance of group A & B are not equal \n ') 
print('########################################################################\n')

print('Step 2: Test the statistics \n\n') 
k = 0

F_test = 2

for L in range(1,12):

    F, p_value= stats.f_oneway(data.iloc[:,L], data.iloc[:,L+11])
    L+=1

    if F > F_test:     
      print("The null hypothesis is rejected ; F = {:.3f} \n".format(F))

    else:
      print("The null hypothesis is accepted ; F = {:.3f} \n".format(F))
      k += 1


if k == 11:
  print('######################################################################## \n')
  print('Step 3: Conclusion \n', 
        'We conclude that the variance of all features between A & B are equal, since all features does not fall into rejection region \n', '########################################################################')
else:
  print('######################################################################## \n')
  print('########################################################################\n\n','Step 3: Conclusion \n', 
        'We conclude that the variance of all features between A & B are NOT equal, since some of the features does fall into rejection region \n')
````

<font size="3"><font color='lime'> +12 Added features : A>B_follower_counts, A<B_following_counts , A>B_follow_ratio,...., A>B retweets_ratio

````python
#0: False, 0.5: Equal, 1: True

def grader(a,b):
  
    if a > b:
      return 1

    elif a==b :
      return 0.5

    else:
      return 0


#Vectorize the function to make it run significantly faster
V_grader = np.vectorize(grader)
````

````python
col_name = ['A>B_follow_ratio',	'A>B_listed_count',	'A>B_mentions_received',	'A>B_retweets_received',	'A>B_mentions_sent',
            'A>B_posts',	'A>B_network_feature_1',	'A>B_network_feature_2',	'A>B_network_feature_3'	,'A>B mentions ratio','A>B retweets ratio']
````

````python
contest_dataset_1['A>B_follower']  = V_grader(contest_dataset_1['A_follower_count'], contest_dataset_1['B_follower_count'])
contest_dataset_1['A<B_following'] = V_grader(contest_dataset_1['B_following_count'], contest_dataset_1['A_following_count'])

col_name = ['A>B_follow_ratio',	'A>B_listed_count',	'A>B_mentions_received',	'A>B_retweets_received',	'A>B_mentions_sent',
            'A>B_posts',	'A>B_network_feature_1',	'A>B_network_feature_2',	'A>B_network_feature_3'	]

#for i in range(3,15):
for i in range(3,12):
  contest_dataset_1[col_name[i-3]] = V_grader(contest_dataset_1.iloc[:,i], contest_dataset_1.iloc[:,i+14])
 
contest_dataset_1['A>B_mentions ratio']  = V_grader(contest_dataset_1['A mentions ratio'],contest_dataset_1['B mentions ratio'])
contest_dataset_1['A>B_retweets ratio'] = V_grader(contest_dataset_1['A retweets ratio'],contest_dataset_1['B retweets ratio'])
````

````python
# Must do K-Fold Cross Validation first

from sklearn.model_selection import cross_val_predict


# This is the criteria I picked for judging the performance of the model
def perf_score(score):
  
  score *= 100

  if score == 100:
    return("Perfect or overfitting?")

  elif score >= 98:            #Ideal score!
    return("Flawless")

  elif score >= 95:
    return("Outstanding")
  
  elif score >= 90:
    return("Excellent")
  
  elif score >= 85:
    return("Great")
  
  elif score >= 80:
    return("Good")
  
  elif score >= 70:
    return("Fair")

  else:
    return("Not acceptable")
````
