## 2) Building the model

```python
#Required Libraries
import numpy as np
import pandas as pd
```

```python
class Simulation:
    def __init__(self):
        self.clock=0                      #Simulation clock
        
        self.num_arrivals=0                 #Total number of arrivals
        
        self.num_in_system = 0
        
        self.t_arrival=self.gen_int_arr()   #Time of next arrival
        
        self.t_departure1=float('inf')      #Departure time from Lavatory 1
        self.t_departure2=float('inf')      #Departure time from Lavatory 2
        self.t_departure3=float('inf')      #Departure time from Lavatory 3
        
        self.dep_sum1=0                     #Sum of service times by Lavatory 1
        self.dep_sum2=0                     #Sum of service times by Lavatory 2
        self.dep_sum3=0                     #Sum of service times by Lavatory 3
        
        self.state_T1=0                     #Current state of Lavatory 1 (binary)
        self.state_T2=0                     #Current state of Lavatory 2 (binary)
        self.state_T3=0                     #Current state of Lavatory 3 (binary)  
        
        self.total_wait_time=0.0            #Total waiting time
        
        self.num_in_q=0                     #Current number in queue
        self.qsum = 0                       #Sum of Queue length of every 15 minutes
        
        self.number_in_queue=0              #Customers who had to wait in line(counter)
        
        self.num_of_departures1=0           #Number of customers served by Lavatory 1  
        self.num_of_departures2=0           #Number of customers served by Lavatory 2
        self.num_of_departures3=0           #Number of customers served by Lavatory 3 
        
        
        self.i=1                                #Iteration every 15 minute
 
        
        
    def time_adv(self):                                                            #(Timing Routine):Determine the Next event
        t_next_event = min(self.t_arrival,self.t_departure1,self.t_departure2,self.t_departure3)  
        self.total_wait_time += (self.num_in_q*(t_next_event-self.clock))
        self.clock = t_next_event        
        
        if min(self.t_arrival,self.t_departure1,self.t_departure2,self.t_departure3) == self.t_arrival :              #
            self.arrival()
        elif min(self.t_arrival,self.t_departure1,self.t_departure2,self.t_departure3) == self.t_departure1:          #
            self.lavatory1()
            
        elif min(self.t_arrival,self.t_departure1,self.t_departure2,self.t_departure3) == self.t_departure2:          #   
            self.lavatory2()
            
        else:
            self.lavatory3()

            
            
            
    def arrival(self):              
        self.num_arrivals += 1
        self.num_in_system += 1

        if self.num_in_q == 0:                                                     #schedule next departure or arrival depending on state of servers
            
            if self.state_T1==1 and self.state_T2==1 and self.state_T3==1:
                self.num_in_q+=1
                self.number_in_queue+=1
                self.t_arrival=self.clock+self.gen_int_arr()
                
                
            elif self.state_T1==0 and self.state_T2==0 and self.state_T3==0:
                
                prob1 = np.random.choice([0,2])
                
                if prob1==1:
                    self.state_T1=1
                    self.dep1= self.gen_service_time_lavatory1()
                    self.dep_sum1 += self.dep1
                    self.t_departure1=self.clock + self.dep1
                    self.t_arrival=self.clock+self.gen_int_arr()

                elif prob1==2:
                    self.state_T2=1
                    self.dep2= self.gen_service_time_lavatory2()
                    self.dep_sum2 += self.dep2
                    self.t_departure2=self.clock + self.dep2
                    self.t_arrival=self.clock+self.gen_int_arr()
                else:
                    self.state_T3=1
                    self.dep3 = self.gen_service_time_lavatory3()
                    self.dep_sum3 += self.dep3
                    self.t_departure3 =self.clock + self.dep3
                    self.t_arrival = self.clock+self.gen_int_arr()
                    
            elif self.state_T1==1 and self.state_T2==0 and self.state_T3==0:
                
                prob2 = np.random.choice([0,1])
                
                if prob2==1:
                    self.state_T2=1
                    self.dep2= self.gen_service_time_lavatory2()
                    self.dep_sum2 += self.dep2
                    self.t_departure2=self.clock + self.dep2
                    self.t_arrival=self.clock+self.gen_int_arr()

                else:
                    self.state_T3=1
                    self.dep3 = self.gen_service_time_lavatory3()
                    self.dep_sum3 += self.dep3
                    self.t_departure3 =self.clock + self.dep3
                    self.t_arrival = self.clock+self.gen_int_arr()
                    
            elif self.state_T1==0 and self.state_T2==1 and self.state_T3==0:
                
                prob3 = np.random.choice([0,1])
                
                if prob3==1:
                    self.state_T1=1
                    self.dep1= self.gen_service_time_lavatory1()
                    self.dep_sum1 += self.dep1
                    self.t_departure1=self.clock + self.dep1
                    self.t_arrival=self.clock+self.gen_int_arr()

                else:
                    self.state_T3=1
                    self.dep3 = self.gen_service_time_lavatory3()
                    self.dep_sum3 += self.dep3
                    self.t_departure3 =self.clock + self.dep3
                    self.t_arrival = self.clock+self.gen_int_arr()
            
            elif self.state_T1==0 and self.state_T2==0 and self.state_T3==1:
                
                prob4 = np.random.choice([0,1])
                
                if prob4==1:
                    self.state_T1=1
                    self.dep1= self.gen_service_time_lavatory1()
                    self.dep_sum1 += self.dep1
                    self.t_departure1=self.clock + self.dep1
                    self.t_arrival=self.clock+self.gen_int_arr()

                else:
                    self.state_T2=1
                    self.dep2 = self.gen_service_time_lavatory2()
                    self.dep_sum2 += self.dep2
                    self.t_departure2 =self.clock + self.dep2
                    self.t_arrival = self.clock+self.gen_int_arr()
            
            elif self.state_T1==1 and self.state_T2==1 and self.state_T3==0:             
                    self.state_T3=1
                    self.dep3 = self.gen_service_time_lavatory3()
                    self.dep_sum3 += self.dep3
                    self.t_departure3 =self.clock + self.dep3
                    self.t_arrival = self.clock+self.gen_int_arr()
        
            elif self.state_T1==0 and self.state_T2==1 and self.state_T3==1:             
                    self.state_T1=1
                    self.dep1 = self.gen_service_time_lavatory1()
                    self.dep_sum1 += self.dep1
                    self.t_departure1 =self.clock + self.dep1
                    self.t_arrival = self.clock+self.gen_int_arr()
                    
            elif self.state_T1==1 and self.state_T2==0 and self.state_T3==1:             
                    self.state_T2=1
                    self.dep2 = self.gen_service_time_lavatory2()
                    self.dep_sum2 += self.dep2
                    self.t_departure2 =self.clock + self.dep2
                    self.t_arrival = self.clock+self.gen_int_arr()  
                    
                    
        else:
            self.num_in_q+=1
            self.number_in_queue+=1                             
            self.t_arrival=self.clock + self.gen_int_arr()
            
            if self.clock > 15*s.i:                
                self.qsum += self.num_in_q
                self.qsum -=1              #Eliminate the one customer that come after 15 minute
                self.i += 1            
        
        
    
    def lavatory1(self):                #departure from server 2 or 3
        self.num_of_departures1 += 1
        if self.num_in_q>0:
            self.dep1= self.gen_service_time_lavatory1()
            self.dep_sum1 += self.dep1
            self.t_departure1=self.clock + self.dep1
            self.num_in_q-=1
        else:
            self.t_departure1=float('inf') 
            self.state_T1=0                  
    
    def lavatory2(self):                #departure from server 1 or 3
        self.num_of_departures2 += 1
        if self.num_in_q>0:
            self.dep2= self.gen_service_time_lavatory2()
            self.dep_sum2 += self.dep2
            self.t_departure2=self.clock + self.dep2
            self.num_in_q-=1
        else:
            self.t_departure2=float('inf')
            self.state_T2=0
            
    def lavatory3(self):                #departure from server 1 or 2
        self.num_of_departures3 += 1
        if self.num_in_q>0:
            self.dep3= self.gen_service_time_lavatory3()
            self.dep_sum3 += self.dep3
            self.t_departure3=self.clock + self.dep3
            self.num_in_q-=1
        else:
            self.t_departure3=float('inf')
            self.state_T3=0
            
    def gen_int_arr(self):                                             #function to generate arrival times using inverse trsform
        return (-np.log(1-(np.random.uniform(low=0.0,high=1.0))) * 3)
    
    def gen_service_time_lavatory1(self):                                #function to generate service time for Lavatory 1 using inverse trnasform
        return (-np.log(1-(np.random.uniform(low=0.0,high=1.0))) * 5)
    
    def gen_service_time_lavatory2(self):                                #function to generate service time for Lavatory 2 using inverse trnasform
        return (-np.log(1-(np.random.uniform(low=0.0,high=1.0))) * 5)
    
    def gen_service_time_lavatory3(self):                                #function to generate service time for Lavatory 3 using inverse trnasform
        return (-np.log(1-(np.random.uniform(low=0.0,high=1.0))) * 5)
```

## 3) Run the simulation

```python
s = Simulation()

df=pd.DataFrame(columns=['Average interarrival time', 'Average service time of of all Lavatories' ,
                         'Utilization of all Lavatories'
                         ,'Total queue length in line','Average queue length every 15 minutes',
                         'Average waiting time']) 
```

```python
for i in range(1000):                                 #The more attempts you add , the more you precise you obtain the data insights
    s.__init__()                                     #Reset
    
    while s.clock <= 720 :
        s.time_adv()                                 #Generate next event
    
    a=pd.Series([s.clock/s.num_arrivals, (s.dep_sum1/s.num_of_departures1+s.dep_sum2/s.num_of_departures2+s.dep_sum3/s.num_of_departures3)/3,
                 (s.dep_sum1/s.clock+s.dep_sum2/s.clock+s.dep_sum3/s.clock)
                 ,s.number_in_queue, s.qsum/s.i,
                    s.total_wait_time/(s.num_of_departures1+s.num_of_departures2+s.num_of_departures3)]
                ,index=df.columns)
    
    df=df.append(a,ignore_index=True) 
```
